```python
import numpy as py
import pandas as pd
import seaborn as sea
from matplotlib import pyplot as plt

sea.set_style('darkgrid')
df_bt = pd.read_csv("C://Users//Librarian//GitHub//Upgrad//Data-Science//Visualizaion_charts/crypto_data/bitcoin_price.csv")
df_eth = pd.read_csv("C://Users//Librarian//GitHub//Upgrad//Data-Science//Visualizaion_charts/crypto_data/ethereum_price.csv")
df_lit =  pd.read_csv("C://Users//Librarian//GitHub//Upgrad//Data-Science//Visualizaion_charts/crypto_data/litecoin_price.csv")

df_bt.columns = 'bt_' + df_bt.columns
df_eth.columns  ='eth_' + df_eth.columns
df_lit.columns = 'lit_' + df_lit.columns



m1 = pd.merge(df_bt, df_eth, how = 'inner',left_on = 'bt_Date', right_on = 'eth_Date')
m2 = pd.merge(m1, df_lit, how = 'inner', left_on = 'bt_Date', right_on = 'lit_Date')

crypto = m2[['bt_Close','eth_Close','lit_Close']]

print(crypto.corr())

sea.pairplot(crypto)

plt.figure(2)
plt.subplot(1,1,1)
sea.heatmap(crypto.corr(),cmap = 'Spectral_r', annot = True)
```

```python
import numpy as py
import pandas as pd
import seaborn as sea
from matplotlib import pyplot as plt


sea.set_style('darkgrid')
market = pd.read_csv("C://Users//Librarian//GitHub//Upgrad//Data-Science//Visualizaion_charts//global_sales_data/market_fact.csv")
prod_dimension = pd.read_csv("C://Users//Librarian//GitHub//Upgrad//Data-Science//Visualizaion_charts//global_sales_data/prod_dimen.csv")
order_dimension = pd.read_csv("C://Users//Librarian//GitHub//Upgrad//Data-Science//Visualizaion_charts//global_sales_data/orders_dimen.csv")
m1 = pd.merge(market,prod_dimension,on = 'Prod_id', how = 'inner')
m2 = pd.merge(m1,order_dimension,how = 'inner', on = 'Ord_id')

m2['Order_Date'] = pd.to_datetime(m2['Order_Date'])


m2['year'] = m2.Order_Date.dt.year
m2['month'] = m2.Order_Date.dt.month

time_df = m2.groupby(['year','month']).mean()['Sales']
#print(time_df)
plt.figure(figsize=(24,16))
sea.tsplot(data=time_df)

```

```python
year_month_sales = pd.pivot_table(m2, values='Sales', index = 'year', columns='month',aggfunc='mean')
print(year_month_sales)
sea.heatmap(year_month_sales,cmap='rainbow')
```
